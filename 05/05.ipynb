Programe las siguientes técnicas de Morfología Matemática Binaria

1. Erosión y dilatación, apertura y cierre para imágenes binarias. En los últimos dos casos puede usar las funciones programadas previamente. Asuma que las imágenes de entrada son binarias y que el elemento estructurante es genérico pero de tamaño fijo de 3x3.

```python
%matplotlib inline
import numpy as np
from tp3 import convolute_image
import matplotlib.pyplot as plt

def erode(im, selement, background=0, figure=1):
    def erode_op(portion, kernel):
        target = portion[kernel==1].flat
        original_value = portion[1,1]
        if (background in target):
            return background
        else:
            return original_value
    
    return convolute_image(im,selement, erode_op)

def dilate(im, selement, background=0, figure=1):
    def erode_op(portion, kernel):
        target = portion[kernel==1].flat
        original_value = portion[1,1]
        if (figure in target):
            return figure
        else:
            return original_value
    
    return convolute_image(im,selement, erode_op)

def neg(im):
    return 1-im

def morph_close(im, selement, background=0, figure=1):
    return erode(
        dilate(im, selement, background, figure),
        selement,
        background,
        figure
    )

def morph_open(im, selement, background=0, figure=1):
    return dilate(
        erode(im, selement, background, figure),
        selement,
        background,
        figure
    )

def morph_open_close(im, selement, background=0, figure=1):
    return morph_close(
        morph_open(
            im,
            selement,
            background,
            figure
        ),
        selement,
        background,
        figure
    )

def morph_inner_border(im, selement, background=1, figure=0):
    return im- erode(im, selement, background, figure)

def morph_outer_border(im, selement, background=1, figure=0):
    return dilate(im, selement, background, figure) - im
```

```python
import imageio
from tp1 import threshold

camera = imageio.imread('imageio:camera.png') / 255
# Threshold at .5
camera = -threshold(camera, norm=True) + 1
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax.set_title('Original Image')
ax.imshow(camera,cmap='gray')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax.set_title('Eroded Image')
ax.imshow(erode(camera, np.ones((3,3))),cmap='gray')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax.set_title('Dilated Image')
ax.imshow(dilate(camera, np.ones((3,3))),cmap='gray')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax.set_title('Opened Image')
ax.imshow(morph_open(camera, np.ones((3,3))),cmap='gray')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax.set_title('Closed Image')
ax.imshow(morph_close(camera, np.ones((3,3))),cmap='gray')
plt.show()
```

2. Reconstrucción Morfológica. Cuando lo requiera utilice las funciones programadas en el punto anterior. 

```python
def cond_dilate(im, selem, marker, background=1, figure=0):
    dilated = dilate(marker, selem, background=background, figure=figure)
    return dilated * im
```

```python
def morph_reconstruct(im, selem, marker, background=1, figure=0):
    return morph_reconstruct_r(im, selem, marker, background=background, figure=figure)

def morph_reconstruct_r(im, selem, marker, background, figure):
    next_iteration = cond_dilate(im, selem, marker, background=background, figure=figure)
    if np.all(next_iteration==marker):
        return marker
    else:
        return morph_reconstruct_r(im, selem,next_iteration,background=background, figure=figure)
```

```python
caracteres = imageio.imread('data/caracteres.bmp') / 255
caracteres = -caracteres[:,:,0]+1 # Keep it in binary
```

```python
marker = np.zeros(caracteres.shape)
marker[100,80] = 1
```

```python
plt.imshow(caracteres, cmap='gray')
```

```python
plt.imshow(marker, cmap='gray')
```

```python
recon = morph_reconstruct(caracteres, np.ones((3,3)), marker, background=0, figure=1)
```
